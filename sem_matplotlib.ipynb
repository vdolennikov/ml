```python
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt

import warnings
warnings.filterwarnings("ignore")
```

```python
fig, axis = plt.subplots(ncols=2, nrows=2, figsize=(15, 8))
fig.suptitle('Plot examples', fontsize=20)

# scatter
mean = [20, 10]
cov = [[1, 0.7], [0.7, 1]]
scatter_data = np.random.multivariate_normal(mean, cov, 1000)
axis[0, 0].scatter(scatter_data[:, 0], 
                   scatter_data[:, 1],
                   edgecolor='black')
axis[0, 0].set_title('Scatter plot example')

# hist
hist_data = np.random.geometric(p=0.1, size=3000)
axis[0, 1].hist(hist_data, bins=40, color='r')
axis[0, 1].set_title('Hist plot example')

# line
line_data = np.linspace(0, 6, 100)
axis[1, 0].plot(line_data, np.log(line_data),
                color='black',
                label='log(x)')
axis[1, 0].plot(line_data, np.log(2*line_data),
                color='r',
                label='log(2x)')
axis[1, 0].legend()
axis[1, 0].set_title('Line plot example')

# bar
bar_data_y = np.random.normal(5, size=14)
bar_data_label = [dt.datetime(2021, 1, 1 + i).strftime('%Y-%m-%d') for i in range(14)]
axis[1, 1].bar(bar_data_label, bar_data_y)
axis[1, 1].set_xticklabels(bar_data_label, rotation = 90)

plt.show()
```
